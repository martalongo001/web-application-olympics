```python
import pandas as pd

file_path = r'C:\Users\Sara\OneDrive\Ambiente de Trabalho\PROJET-BASES-DADOS\ollympics\athlete_events.csv'

# Read the CSV file
df = pd.read_csv(file_path)
df.head()
```

```python
import pandas as pd

#novo_nome = {'ID': 'idAtletas'}

#df = df.rename(columns = novo_nome)

df['idAtletas'] = pd.factorize(df['Name'].astype(str) + '_' + df['Age'].astype(str))[0]
df['idEquipas'] = pd.factorize(df['Team'].astype(str) +'_' + df['NOC'].astype(str))[0]
df['idCategorias'] = pd.factorize(df['Event'])[0]
df['idEventos'] = pd.factorize(df['Games'].astype(str) + '_' + df['City'].astype(str))[0]
df['idModalidades'] = pd.factorize(df['Sport'])[0]


df.to_csv('athlete_events.csv', index=False)
df
```

```python
#Atletas

colunas_atletas = ['idAtletas', 'Name', 'Sex', 'Age', 'Height', 'Weight', 'idEquipas']

atletas_df = df[colunas_atletas]

atletas_df = atletas_df.drop_duplicates(subset='idAtletas')

atletas_df.to_csv('C:\\Users\\Sara\OneDrive\\Ambiente de Trabalho\\PROJET-BASES-DADOS\\ollympics\\Atletas.csv', index=False)
atletas_df
```

```python
#Equipas

colunas_equipas = ['idEquipas','Team', 'NOC']

equipas_df = df[colunas_equipas]

equipas_df = equipas_df.drop_duplicates()

equipas_df.to_csv('C:\\Users\\Sara\OneDrive\\Ambiente de Trabalho\\PROJET-BASES-DADOS\\ollympics\\Equipas.csv', index=False)
equipas_df

```

```python
#Eventos
colunas_eventos = ['idEventos', 'Year', 'Season', 'City']

eventos_df = df[colunas_eventos]
eventos_df = eventos_df.drop_duplicates()

eventos_df.to_csv('C:\\Users\\Sara\OneDrive\\Ambiente de Trabalho\\PROJET-BASES-DADOS\\ollympics\\Eventos.csv', index=False)
eventos_df
```

```python
#Modalidades

colunas_modalidades = ['idModalidades', 'Sport']

modalidades_df = df[colunas_modalidades]
modalidades_df = modalidades_df.drop_duplicates()

modalidades_df.to_csv('C:\\Users\\Sara\OneDrive\\Ambiente de Trabalho\\PROJET-BASES-DADOS\\ollympics\\Modalidades.csv', index=False)
modalidades_df.head()
```

```python
#Categorias
colunas_categorias =['idCategorias', 'Event', 'idModalidades']

categorias_df = df[colunas_categorias]
categorias_df = categorias_df.drop_duplicates()

categorias_df.to_csv('C:\\Users\\Sara\OneDrive\\Ambiente de Trabalho\\PROJET-BASES-DADOS\\ollympics\\Categorias.csv', index=False)
categorias_df.head()
```

```python
#Participacoes
colunas_participacoes = ['idAtletas', 'idEventos', 'idCategorias', 'Medal']

participacoes_df = df[colunas_participacoes]

participacoes_df.to_csv('C:\\Users\\Sara\OneDrive\\Ambiente de Trabalho\\PROJET-BASES-DADOS\\ollympics\\Participacoes.csv', index=False)
participacoes_df.head()
```
